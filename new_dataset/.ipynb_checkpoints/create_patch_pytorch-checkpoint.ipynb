```python
import os
import cv2
import numpy as np

path_label = '/disk2/dataset_0724_for_pytorch_label/'
path = '/disk2/dataset_0724_for_pytorch/'    
label_list = os.listdir('/disk2/dataset_0724/label/')
useful_list = []
for n in label_list:
    if n[13] == 'T':
        useful_list.append(n)
    
n = 0
step_size = 800
img_scale = 800

for i in useful_list[0:-10]:
    label_path = '/disk2/dataset_0724/label/' + i
    img_path = '/disk2/dataset_0724/picture/img' + i[5:]
    
    label = cv2.imread(label_path, 1)
    img = cv2.imread(img_path, 1)
    
    for x in xrange(0, img.shape[0], step_size):
        for y in xrange(0, img.shape[1], step_size):
            if (x+step_size > img.shape[0]) or (y+step_size > img.shape[1]):
                break
            img_window = img[x:x+img_scale, y:y+img_scale]
            label_window = label[x:x+img_scale, y:y+img_scale]
            if np.max(label_window) == 255:
                cv2.imwrite(path+'/1/'+'{0}.jpg'.format(n), img_window)
                cv2.imwrite(path_label+'/1/' +'{0}.jpg'.format(n), label_window)      
            if np.max(label_window) == 0:
                cv2.imwrite(path+'/0/'+'{0}.jpg'.format(n), img_window)
                cv2.imwrite(path_label+'/0/' +'{0}.jpg'.format(n), label_window)    
            n+=1
            print(np.max(label_window))
```
